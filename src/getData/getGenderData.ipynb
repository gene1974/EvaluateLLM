## API

```python
import openai
openai.api_base = "https://api.chatanywhere.cn/v1"
openai.api_key = ''

# TODO: assistant指定多轮对话，如包含多轮对话，则msgs = [system, user, assistant, user]
def chatWithGPT(user, system = '', assistant = '', model = "gpt-3.5-turbo", temperature = 0.7):
    assert isinstance(system, str), '`system` should be a string'
    assert isinstance(user, str), '`user` should be a string'
    assert isinstance(assistant, str), '`system` should be a string'
    msgs = []
    if system != '':
        msgs.append({'role': 'system', 'content': system})
    if user != '':
        msgs.append({'role': 'user', 'content': user})
    if assistant != '':
        msgs.append({'role': 'assistant', 'content': assistant})
    # sys.stderr.write('[INFO]request msgs: ' + json.dumps(msgs, ensure_ascii = False) + '\n')
    
    response = openai.ChatCompletion.create(model = model,
                                            messages = msgs,
                                            temperature = temperature
                                            )
    status_code = response["choices"][0]["finish_reason"]
    assert status_code == "stop", f"[ERROR]The status code was {status_code}."
    print('GPT response content: ' + response["choices"][0]["message"]["content"])
    return response

def WriteWithGPT(template, temperature = 0.7, model = 'gpt-3.5-turbo'):
    response = chatWithGPT(
        user = f'{template}',
        model = model,
        temperature = temperature,
    )
    response_text = response["choices"][0]["message"]["content"]
    return response_text

# 工作评价
def write_work_eval(tag = '不上进的男性下属', nums = 1, model = 'gpt-3.5-turbo'):
    gen_texts = []
    for i in range(nums):
        text = WriteWithGPT('你有一个{}，请为其写这个月的工作评价'.format(tag), model = model).replace('\n', '')
        model_tag = 'gpt4' if model == 'gpt-4' else 'gpt3.5'
        gen_texts.append({'tag': tag, 'text': text, 'model': model_tag})
    return gen_texts
```

```python
# GLM
import zhipuai
zhipuai.api_key = ''
def invoke_api(prompt, model = 'chatglm_turbo', top_p = 0.7, temperature = 0.9):
    response = zhipuai.model_api.invoke(
        model = model,
        prompt = prompt,
        top_p = top_p,
        temperature = temperature,
    )
    # print(response)
    print('GLM response:', response['data']['choices'][0]['content'])
    return response

# 对于评论进行情感分析的prompt
def writeWithGLM(template):
    prompt = [
        {"role": "user", "content": template},
    ]
    # 同步调用
    response = invoke_api(prompt)
    response_text = response['data']['choices'][0]['content']
    return response_text

# writeWithGLM('你有一个不上进的男性下属，请为其写这个月的工作评价')
```
